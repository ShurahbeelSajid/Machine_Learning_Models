#**Sci Kit Learn Model Training**

*Goal: Determine whether a data is soluble in water or solvent*

##Loading Data

Link of CSV File: [Dataset](https://github.com/dataprofessor/data/blob/master/delaney_solubility_with_descriptors.csv)

```python
import pandas as pd
file_path = 'C:/Users/shura/Downloads/delaney_solubility_with_descriptors.csv'
df = pd.read_csv(file_path)
print(df)
```

##Data Preparation

### Data Separation as X and y

```python
y = df['logS']
print(y)
```

```python
X = df.drop('logS', axis=1)
print(X)
```

###Data Spliting

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python
X_train
```

```python
X_test
```

## **Building a Model**

### Linear Regression

####**Training the Model**

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

####**Applying the Model to Calculate Predictions**

```python
y_lr_train_pred = lr.predict(X_train)
y_lr_test_pred = lr.predict(X_test)
```

```python
print(y_lr_train_pred)
```

```python
print(y_lr_test_pred)
```

#### **Evaluate Model Performance**

```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
print('LR MSE (Train): ', lr_train_mse)
print('LR R2 (Train): ', lr_test_r2)
print('LR MSE (Test): ', lr_test_mse)
print('LR MSE (Test): ', lr_test_r2)
```

```python
lr_results = pd.DataFrame(['Linear Regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns = ('Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2')
```

```python
print(lr_results)
```

### Random Forest

#### Training the Model

```python
from sklearn.ensemble import RandomForestRegressor #We have two versions of Random Forest as Random Forest Regressor and Random Forest Classifier

rf = RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(X_train, y_train)
```

#### Applying the Model to Calculate Predictions

```python
y_rf_train_pred = rf.predict(X_train)
y_rf_test_pred = rf.predict(X_test)
```

#### Evaluate Model Performance

```python
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
print('RF MSE (Train): ',rf_train_mse)
print('RF R2 (Train): ', rf_test_r2)
print('RF MSE (Test): ', rf_test_mse)
print('RF MSE (Test): ', rf_test_r2)
```

```python
rf_results = pd.DataFrame(['Random Forest', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns = ('Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2')
```

```python
print(rf_results)
```

##Model Comparison

```python
df_models = pd.concat([lr_results, rf_results], axis=0)
print(df_models)
```

```python
df_models.reset_index(drop=True)
```
